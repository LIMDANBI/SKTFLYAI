```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

# X-Y 좌표로 이루어진 데이터 불러오기 

```python
df = pd.read_csv('./datasets/cluster.csv')
df
```

# 불러온 데이터 시각화

```python
sns.scatterplot(x='X', y='Y', data=df)
plt.show()
```

# 모델링
## K 값 변경 (1~5)

```python
for n in range(1, 6):
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(df)
    df_copy = df.copy()
    df_copy['cluster'] = kmeans.labels_
    sns.scatterplot(x='X', y='Y', data=df_copy, hue='cluster', palette='Set1')
    plt.show()
```

```python
df_copy
```

# 임의의 데이터 넣어보기

```python
import numpy as np
tmp = np.array([[1.17, 1.36]])
```

```python
pred_cluster = kmeans.predict(tmp)[0]
print("x = {}, y ={} 데이터는 cluster {} 입니다.".format(tmp[0][0], tmp[0][1], pred_cluster))
```
