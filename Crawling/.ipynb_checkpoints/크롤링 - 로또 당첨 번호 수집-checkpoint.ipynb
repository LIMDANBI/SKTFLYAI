```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
res = requests.get('https://dhlottery.co.kr/common.do?method=main')
res
```

```python
soup = bs(res.content, 'html.parser')
soup
```

```python
result = soup.select('.ball_645') # class = . / id = #
result
```

```python
for num in result:
    print(num.text,end="번 ")
```
