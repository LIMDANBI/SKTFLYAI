## 데이터 불러오기

pregnant: 과거 임신 횟수 <br>
plasma: 포도당 부하 검사 2시간 후 공복 혈당 농도(mm Hg) <br>
pressure: 확장기 혈압(mm Hg) <br>
thickness: 삼두근 피부 주름 두께 (mm) <br>
insulin: 혈청 인슐린(mu U/ml) <br>
bmi: 체질량지수(체중(kg)/키(m))^2 <br>
pedigree: 당뇨병 가족력 <br>
age: 나이 <br>
<br>
diabetes: 당뇨 여부(0 또는 1) <br>

## feature - label 분리

```python
X = df.iloc[:, :8]
y = df.iloc[:, 8]
```

## 모델링

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu', name='Dense1'))
model.add(Dense(8, activation='relu', name='Dense2'))
model.add(Dense(1, activation='sigmoid', name='Dense3'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
hisotry = model.fit(X, y, epochs=500, batch_size=16)
```
