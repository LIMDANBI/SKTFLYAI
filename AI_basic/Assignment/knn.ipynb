```python
def KNNClassifier(data, mydata, labels, K, distname):
    cost = []
    for sample, label in zip(data, labels):
        if distname == 'manhattan':
            c = np.sum(abs(sample-mydata)) # manhattan distance
        elif distname == 'euclidean':
            c = np.sum(np.sqrt(np.power((sample-mydata),2))) # euclidean distance
        cost.append([c, label])
    cost.sort()
    res_labels = np.array(cost[:K])[:,1]
    return st.multimode(res_labels)[0]
```
