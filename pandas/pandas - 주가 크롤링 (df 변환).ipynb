```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
finance_url = 'https://finance.naver.com/'
res = requests.get(finance_url)
res
```

```python
soup = bs(res.content, "html.parser")
soup
```

```python
name = soup.select('#_topItems1 > tr > th')
up_down = soup.select('#_topItems1 > tr > td')
```

```python
name
```

```python
up_down
```

```python
name_li = []
for n in name:
    name_li.append(n.text)
```

```python
price = []
updown = []
ratio = []
for i in range(len(up_down)):
    if i%3==0:
        price.append(up_down[i].text)
    elif i%3==1:
        updown.append(up_down[i].text)
    else:
        ratio.append(up_down[i].text)
```

```python
price
```

```python
updown
```

```python
ratio
```

# pandas dataframe 만들기

```python
import pandas as pd
```

```python
df = pd.DataFrame([price, updown, ratio], index = ['주가','등락', '등락비율'], columns=name_li)
df
```
