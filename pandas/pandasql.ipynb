```python
from pandasql import sqldf # dataframe에 SQL 이용

dfsql = lambda q:sqldf(q, globals()) # globals()로 환경을 일괄적으로 지정
result = dfsql('select name,episodes,type from df where type="Movie" order by episodes desc')
result
# result.to_json('./sample/sample.json', orient='table') # json parser 에서 쉽게 분석 가능 ~!
```
