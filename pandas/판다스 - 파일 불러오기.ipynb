## JSON
### http://www.gimhae.go.kr/openapi/tour/festival.do

```python
df_json = pd.read_json('http://www.gimhae.go.kr/openapi/tour/festival.do')
df_json
```

```python
df_json['results'][0]
```

```python
idx = []
name = []
category = []
area = []
copy = []
manage = []
phone = []
for i in range(len(df_json['results'][:])):
    idx.append(df_json['results'][i]['idx'])
    name.append(df_json['results'][i]['name'])
    category.append(df_json['results'][i]['category'])
    area.append(df_json['results'][i]['area'])
    copy.append(df_json['results'][i]['copy'])
    manage.append(df_json['results'][i]['manage'])
    phone.append(df_json['results'][i]['phone'])
```

```python
df_json['idx'] = idx
df_json['name'] = name
df_json['category'] = category
df_json['area'] = area
df_json['copy'] = copy
df_json['manage'] = manage
df_json['phone'] = phone
```

```python
df_json.drop('results', axis=1, inplace=True)
df_json
```
