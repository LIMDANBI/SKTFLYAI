```python
import newspaper
import matplotlib.pyplot as plt
from wordcloud import WordCloud

font_path = './sample/NanumGothic.ttf'
wordcloud = WordCloud(font_path=font_path, background_color="white", width=400, height=400)


# 기사 수집
link = 'https://n.news.naver.com/mnews/article/009/0004991987?sid=105'
article = newspaper.Article(link, language="ko")
article.download()
article.parse()
# print(article.text)

# 분석
wordcloud = wordcloud.generate(article.text)
fig = plt.figure(figsize=(6,6))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
fig.savefig("worldcloud.png")
```

# dataframe 시각화

```python
import os
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
csv = os.path.join('./sample/anime_stock_returns.csv')
df = pd.read_csv(csv, index_col=0, parse_dates=['Date'])
df.head()
```

```python
plt.style.use('ggplot')
fig  = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)
ax.plot(df.index, df['TOEI ANIMATION'], label='TOEI ANIMATION')
ax.plot(df.index, df['IG Port'], label='IG Port')
ax.legend()
ax.set_ylabel('Juga')
ax.set_xlabel('Day')
plt.show()
```

# 공공데이터 시각화

```python
df = pd.read_csv('./sample/행정안전부_재난피해 심리상담실적_12_31_2020.csv', encoding='CP949')
df
```

```python
import matplotlib.pyplot as plt

if os.name == 'posix':
    plt.rc("font", family="AppleGothic")

else:
    plt.rc("font", family="Malgun Gothic")
```

```python
plt.style.use('ggplot')
fig  = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(121)
ax1.set_title("지역별 20대 재난피해 심리상담실적")
ax2 = fig.add_subplot(122)
ax2.set_title("지역별 30대 재난피해 심리상담실적")
ax1.bar(df['구분'], df['연령(20-29세)'])
ax2.bar(df['구분'], df['연령(30-39세)'])
plt.show()
```
