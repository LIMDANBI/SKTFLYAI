```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

survived : 생존=1, 죽음=0 <br>
pclass : 승객 등급. 1등급=1, 2등급=2, 3등급=3 <br>
sibsp : 함께 탑승한 형제 또는 배우자 수 <br>
parch : 함께 탑승한 부모 또는 자녀 수 <br>
ticket : 티켓 번호 <br>
cabin : 선실 번호 <br>
embarked : 탑승장소 S=Southhampton, C=Cherbourg, Q=Queenstown

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# 회귀선이 있는 산점도

```python
sns.set_style('whitegrid') # # 스타일 테마 설정 (5가지: darkgrid, whitegrid, dark, white, ticks)

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

sns.regplot(x='age', y='fare', data=titanic, ax=ax1)
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=False) # 선형회귀선 미표시

plt.show()
```

# 히스토그램/커널 밀도 그래프

```python
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

sns.distplot(x=titanic['fare'], ax=ax1) 
sns.kdeplot(x='fare', data=titanic, ax=ax2) 
sns.histplot(x='fare', data=titanic, ax=ax3)

ax1.set_title('titanic fare - distplot')
ax2.set_title('titanic fare - kedplot')
ax3.set_title('titanic fare - histplot')

plt.show()
```

# Heat Map

```python
table = titanic.pivot_table(index=['sex'], columns=['class'], aggfunc='size') # 데이터 크기에 맞도록
table
```

```python
sns.heatmap(table, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5, cbar=True)
plt.show()
```

# 범주형 데이터의 산점도

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

sns.stripplot(x="class", y="age", data=titanic, ax=ax1, hue='sex') # 데이터 분산 미고려
sns.swarmplot(x="class", y="age", data=titanic, ax=ax2, hue='sex') # 데이터 분산 고려 (중복 X) 

plt.show()
```

# 막대 그래프

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

sns.barplot(x='sex', y='survived', data=titanic, ax=ax1)
sns.barplot(x='sex', y='survived', hue='class', data=titanic, ax=ax2)
sns.barplot(x='sex', y='survived', hue='class', dodge=False, data=titanic, ax=ax3) #  복수의 데이터를 독립적인 막대 그래프로 나란히 표현 x 

ax1.set_title('titanic survived - sex')
ax2.set_title('titanic survived - sex/class')
ax3.set_title('titanic survived - sex/class(stacked)')

plt.show()
```

# 빈도 그래프 

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

sns.countplot(x='class', palette='Set1', data=titanic, ax=ax1)
sns.countplot(x='class', hue='who', palette='Set2', data=titanic, ax=ax2)
sns.countplot(x='class', hue='who', palette='Set3', dodge=False, data=titanic, ax=ax3)

ax1.set_title('titanic class')
ax2.set_title('titanic class - who')
ax3.set_title('titanic class - who(stacked)')

plt.show()
```

# 박스 플롯/바이올린 그래프

```python
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

sns.boxplot(x='alive', y='age', data=titanic, ax=ax1)
sns.boxplot(x='alive', y='age', hue='sex', data=titanic, ax=ax2)
sns.violinplot(x='alive', y='age', data=titanic, ax=ax3)
sns.violinplot(x='alive', y='age',  hue='sex', data=titanic, ax=ax4)

plt.show()
```

# 조인트 그래프

jointplot() 함수는 산점도를 기본으로 표시하고, x – y 축에 각 변수에 대한히스토그램을 동시에 보여준다. 따라서 두 변수의 관계와 데이터가 분산되어 있는정도를 한눈에 파악하기 좋다. 

```python
# 조인트 그래프 - 산점도(기본값)
j1 = sns.jointplot(x='fare', y='age', data=titanic)
j1.fig.suptitle('titanic fare - scatter', size=15)
plt.show()
```

```python
# 조인트 그래프 - 회귀선
j2 = sns.jointplot(x='fare', y='age', kind='reg', data=titanic)
j2.fig.suptitle('titanic fare-reg ')
```

```python
# 조인트 그래프 - 육각 그래프 
j3 = sns.jointplot(x='fare', y='age', kind='hex', data=titanic)
j3.fig.suptitle('titanic fare - hex', size=15)
plt.show()
```

```python
# 조인트 그래프 - 커럴 밀집 그래프
j4 = sns.jointplot(x='fare', y='age', kind='kde', data=titanic)
j4.fig.suptitle('titanic fare - kde', size=15)
plt.show()
```
